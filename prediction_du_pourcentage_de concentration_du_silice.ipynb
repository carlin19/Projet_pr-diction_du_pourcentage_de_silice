## Prépareration des données pour l'entrainement

```python
df_iron_target = mining_df['% Silica Concentrate']
df_iron = mining_df.drop(columns = '% Silica Concentrate')
```

```python
df_iron.shape
```

```python
df_iron_target.shape
```

```python
df_iron = np.array(df_iron)
df_iron_target = np.array(df_iron_target)
```

```python
df_iron_target = df_iron_target.reshape(-1,1)
```

```python
df_iron.shape
```

```python
df_iron_target.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_x = StandardScaler()
X = scaler_x.fit_transform(df_iron)
scaler_y = StandardScaler()
y = scaler_y.fit_transform(df_iron_target)

```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 0)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

## Entrainement et Evaluation du modèle linéaire multiple

```python
from sklearn.linear_model import LinearRegression
Multiple_LinearRegression = LinearRegression()
Multiple_LinearRegression.fit(X_train, y_train)
```

```python
accuracy_Multiple_LinearRegression = Multiple_LinearRegression.score(X_test, y_test)
accuracy_Multiple_LinearRegression
```

## Entrainement et Evaluation du modèle d'arbre de décision pour la régression

```python
from sklearn.tree import DecisionTreeRegressor
DecisionTree_model = DecisionTreeRegressor(random_state=0)
DecisionTree_model.fit(X_train, y_train)
```

```python
accuracy_DecisionTree = DecisionTree_model.score(X_test, y_test)
accuracy_DecisionTree
```

## Entrainement et Evaluation du modèle de foret aléatoire pour la régression

```python
from sklearn.ensemble import RandomForestRegressor

RandomForestRegressor_model = RandomForestRegressor( n_estimators = 10, max_depth = 10, random_state=0)
RandomForestRegressor_model.fit(X_train, y_train)
```

```python
accuracy_RandomForestRegressor = RandomForestRegressor_model.score(X_test, y_test)
accuracy_RandomForestRegressor
```
